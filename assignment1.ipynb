### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.

````python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
#a) Filter out invalid transactions where price is zero or quantity is negative

filtered_transactions = transactions.copy()
for tran in transactions:
  if tran.get('price') == 0 or tran.get('quantity') < 0:
    filtered_transactions.remove(tran)

print(filtered_transactions)

#b) For each remaining transaction, compute a new field total = price * quantity.
for fil in filtered_transactions:
  fil['total'] = fil['price'] * fil['quantity']
print(filtered_transactions)

#c) Group the valid transactions by region and compute the total revenue per region. Store the result in a dictionary.
region_revenue = {}
for rev in filtered_transactions:
  region = rev.get('region')
  total_rev = rev.get('total')
  if region not in region_revenue:
    region_revenue[region] = 0
  region_revenue[region] += total_rev

print(region_revenue)

#d) Print the region with the highest total revenue and its value using an f-string.
highest_rev_region = max(region_revenue, key = region_revenue.get)
print(f"{highest_rev_region} is the region with the highest total revenue of {region_revenue[highest_rev_region]}.")
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
def summarise_sales (sale):
  #create a dictionary
  sales_figure = {}

  #calculate sum of all daily sales
  total = sum(sale)
  sales_figure['total'] = total

  #calculate avg of daily sales value, rounded to two decimals
  average = round(total/len(sale),2)
  sales_figure['average'] = average

  #find the index (starting from 1) of the day with the highest sales
  #can also use max() instead of for loop here
  best = sale[0]
  for s in sale:
    if s > best:
      best = s
  sales_figure['best_day'] = sale.index(best)+1

  #find the index (starting from 1) of the day with the lowest sales
  #can also use min() instead of for loop here
  worst = sale[0]
  for s in sale:
    if s < worst:
      worst = s
  sales_figure['worst_day'] = sale.index(worst)+1

  #calculate the number of days where sales exceeded the average
  count = 0
  for s in sale:
    if s > average:
      count+=1
  sales_figure['above_average_count'] = count

  return (sales_figure)

sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

#call and print out the function
print(summarise_sales(sales))
````

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python

    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

#a) A method add_asset(ticker, shares, price_per_share) that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).
    def add_asset(self, ticker, shares, price_per_share):
      if ticker not in self.holdings:
        self.holdings[ticker] = {'shares': shares, 'price_per_share': price_per_share}
      else:
        self.holdings[ticker]['shares']+= shares
        self.holdings[ticker]['price_per_share'] = price_per_share

#b) A method total_value() that returns the total market value of the portfolio.
    def total_value (self):
      total = 0
      for t in self.holdings:
        total+= self.holdings[t]['shares'] * self.holdings[t]['price_per_share']
      return total

#c) A method most_valuable() that returns the ticker of the asset with the highest total value (shares × price).
    def most_valuable(self):
      max_ticker = max(self.holdings, key = lambda t: self.holdings[t]['shares'] * self.holdings[t]['price_per_share'])
      return max_ticker
#d) A __str__ method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.
    def __str__(self):
      return f"Owner: {self.owner}, Assets: {len(self.holdings)}, Total value: {self.total_value()}"

#Create an instance, add at least three assets, and demonstrate all methods with printed output.
portfolio_1 =  Portfolio('Tanti')
portfolio_1.add_asset('AGX',5000,566.62)
portfolio_1.add_asset('U',10000,19.45)
portfolio_1.add_asset('ADMA',2000,9.25)

print(f"The total market value of the portfolio is: {portfolio_1.total_value()}.")
print(f"The ticker of the asset with the highest total value is: {portfolio_1.most_valuable()}")
print(f"Portfolio summary:\n {portfolio_1}")
````

````python
import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

#a) Compute and print the total annual revenue for each year (i.e., row sums). Use the axis parameter.
years = [2022, 2023, 2024]
row_sums = np.sum(revenue, axis = 1)
for i in range(len(years)):
  print(f"{years[i]}: {row_sums[i]}")

#b) Compute the mean quarterly revenue across all years (i.e., column means).
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
mean_quarterly = np.mean(revenue, axis = 0)
for i in range(len(quarters)):
  print(f"{quarters[i]}:{mean_quarterly[i]:.2f}")

#c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array [1.05, 1.03, 1.08, 1.02]. Print the adjusted values.
adjustment = np.array([1.05, 1.03, 1.08, 1.02])
adjusted = revenue[-1] * adjustment
print(f"Adjusted values:\n {adjusted}")

#d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).
exceeding = revenue[revenue>250]
print(f"Quarterly revenue values that exceed 250 (thousands of euros): \n {exceeding}")
````

### Task 8: More Pandas (1p)

A company records daily website traffic over four weeks. Generate the data by running the following script first:

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

````python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

#a) Create a DataFrame with columns: date, visitors, conversion_rate, and avg_order_value. Set date as the index.
#a1/2 create a dictionary
data = {
    'date': days,
    'visitors': visitors,
    'conversion_rate': conversion_rate,
    'avg_order_value': avg_order_value
}

#a2/2 create a dataframe

data_df = pd.DataFrame(data).set_index('date')

#b) Compute two new columns: conversions (= visitors × conversion_rate, rounded to the nearest integer) and daily_revenue (= conversions × avg_order_value).
data_df['conversions'] = (data_df['visitors'] * data_df['conversion_rate']).round(0).astype(int)
data_df['daily_revenue'] = data_df['conversions'] * data_df['avg_order_value']

print(data_df)
````
